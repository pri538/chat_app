```python
X_norm = df[['cpu_name','cpu_speed','Ram','ssd','gpu_brand','resolution_width','resolution_height','TypeName','gpu_name']].values
y_norm = df['Price'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_norm, y_norm, test_size=0.3, random_state=101)
```

```python
standart_scaler = preprocessing.StandardScaler()

X_train = standart_scaler.fit_transform(X_train)
X_test = standart_scaler.transform(X_test)
```

```python
def display_scores(scores):
    print("Scores: {0}\nMean: {1:.3f}\nStd: {2:.3f}".format(scores, np.mean(scores), np.std(scores)))
```

```python
LR = LinearRegression()

# k-fold CV
scores = cross_val_score(LR, X_train, y_train, scoring='r2', cv=5)
display_scores(scores)
```

```python
LR.fit(X_train,y_train)
y_pred = LR.predict(X_test)
#  model evaluation
f = plt.figure(figsize=(12,8))
ax = f.add_subplot(121)
sns.scatterplot(x = y_test,y = y_pred,ax=ax,color='r')
ax.set_title('Check for Linearity:\n Actual Vs Predicted value')

ax = f.add_subplot(122)
sns.scatterplot(y = (y_test - y_pred), x= y_pred, ax = ax,color='r')
ax.set_title('Check for Homoscedasticity: \nResidual Vs Predicted');
```
